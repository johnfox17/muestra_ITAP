```python
#En esta parte se incluyen las librerias que se utilizaran

import pandas as pd
import matplotlib.pyplot as plt

import plotly.plotly as py
import plotly.tools as tls
import pylab
```

```python
#Aqui se lee el archivo csv que se obtuvo de http://www.beta.inegi.org.mx/proyectos/enchogares/especiales/enadid/2014/default.html

df = pd.read_csv('tfec_hemb.csv')
```

```python
#El metodo .head() te muestra los primeros renglones para que te des una idea de los elementos de la base de datos

df.head()
```

```python
#el archivo tfec_heb.csv esta compuesta por muchas columnas pero solo nos enfocaremos en nivel de escolaridad

df2 = df['niv']
```

```python
#En esta parte se cuenta la frecuencia de cada uno de los datos y se almacena en una serie de pandas llamada df3
df3 =pd.Series.value_counts(df2)
```

```python
#Aqui se ve el tipo de objeto resultante del paso anterior
type(df3)
```

```python
#Se vuelve a convertir en una base de datos de Pandas
df4 = pd.DataFrame(df3)
print(df4)
```

```python
#Se analizan los valores de los indices. El significado de cada numero se encuentra en la documentacion anexa a los datos
df4.index.values
```

```python
#No queremos que nuestra base de datos pandas tenga como indices los valores de escolaridad, asi que los reiniciamos
df5 = pd.DataFrame.reset_index(df4)
print(df5)
```

```python
#Aqui se construye una grafica para tener una mejor manera de presentar los datos. Una manera visual.
#Porque creen que se ve tan raro?

barlist = plt.bar(df5['index'],df5['niv'])

barlist[0].set_color([.1,.2,.5])
barlist[1].set_color([.6,.2,.8])
barlist[2].set_color([.3,.1,.2])
barlist[3].set_color([.5,.8,.4])
barlist[4].set_color([.2,.5,.3])
barlist[5].set_color([.4,.7,.1])
barlist[6].set_color([.8,.6,.7])
barlist[7].set_color([.1,.3,.6])
barlist[8].set_color([.7,.9,.7])
barlist[9].set_color([.5,.1,.3])
plt.show()
```

```python
#Aqui se elimina el dato realmente innecesario del paso anterior
df6 = df5.drop(df5.index[[12]])
```

```python
#Se vuelve a construir la grafica

barlist = plt.bar(df6['index'],df6['niv'])

barlist[0].set_color([.1,.2,.5])
barlist[1].set_color([.6,.2,.8])
barlist[2].set_color([.3,.1,.2])
barlist[3].set_color([.5,.8,.4])
barlist[4].set_color([.2,.5,.3])
barlist[5].set_color([.4,.7,.1])
barlist[6].set_color([.8,.6,.7])
barlist[7].set_color([.1,.3,.6])
barlist[8].set_color([.7,.9,.7])
pylab.xlim([0,15])
plt.xlabel('Escolaridad')
plt.ylabel('Frecuencia')
plt.show()
```
